### Import Output Data

```python
submission_lstm = pd.read_csv('/kaggle/input/submissionslstmlightgbm/submission_lstm.csv')
submission_lightgbm = pd.read_csv('/kaggle/input/submissionslstmlightgbm/submission_lightgbm.csv')
submission = pd.read_csv('/kaggle/input/m5-forecasting-accuracy/sample_submission.csv')
```

```python
submission.shape
```

### Weighted Average Computation

```python
lstm_fac = 0.9
lightgbm_fac = 1.0 - lstm_fac
while (lstm_fac >= 0.1):
    for i in submission.columns:
        if i != 'id':
            submission[i] = submission_lstm[i] * lstm_fac + submission_lightgbm[i] * lightgbm_fac
    submission.to_csv('submission_' + str(int(lstm_fac*10)) + '.csv')
    lstm_fac = lstm_fac - 0.1
    lightgbm_fac = 1.0 - lstm_fac
    print(lstm_fac)
    print(lightgbm_fac)
```

### Computing 0.95 factor, not used in paper

```python
lstm_fac = 0.05
lightgbm_fac = 0.95
for i in submission.columns:
    if i != 'id':
        submission[i] = submission_lstm[i] * lstm_fac + submission_lightgbm[i] * lightgbm_fac
submission.to_csv('submission_05.csv')
```
